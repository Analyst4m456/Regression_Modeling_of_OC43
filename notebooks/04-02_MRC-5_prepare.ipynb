```python
import os
import sys
import mygene
import pandas as pd

if os.getcwd().endswith("notebooks"):
    os.chdir("..")

sys.path.append(os.path.abspath("./"))

print(f"Current work directory: {os.getcwd()}")
```

```python
df = pd.read_csv('./CSV/MRC-5_raw_counts.csv')
```

```python
df
```

```python
mg = mygene.MyGeneInfo()

ensembl_ids = df['Geneid'].unique().tolist()

# Ensembl gene ID → Gene symbol
res = mg.querymany(ensembl_ids,
                   scopes='ensembl.gene',
                   fields='symbol',
                   species='human')

id2symbol = {
    r['query']: r.get('symbol')
    for r in res
    if not r.get('notfound', False)
}

df['Gene_symbol'] = df['Geneid'].map(id2symbol)
```

```python
df
```

```python
mask = df['Gene_symbol'].isna() & ~df['Geneid'].str.startswith('ENSG', na=False)
df.loc[mask, 'Gene_symbol'] = df.loc[mask, 'Geneid']
```

```python
df = df.dropna(subset=['Gene_symbol'])
```

```python
cols = ['Gene_symbol'] + [col for col in df.columns if col not in ['Gene_symbol', 'Geneid']]

df = df[cols]

df.head()
```

```python
df.to_csv("./CSV/Bulk-seq/GSE252692_converted.csv", index=False)
```

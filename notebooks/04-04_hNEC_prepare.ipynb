```python
import os
import sys
import mygene
import pandas as pd

if os.getcwd().endswith("notebooks"):
    os.chdir("..")

sys.path.append(os.path.abspath("./"))

print(f"Current work directory: {os.getcwd()}")
```

```python
df = pd.read_table('./Data/Bulk-seq_data/GSE238079_HCoV_nae_PAPER_raw_gene_counts.txt')
```

```python
mg = mygene.MyGeneInfo()

ensembl_ids = df['Gene'].unique().tolist()

# Ensembl gene ID â†’ Gene symbol
res = mg.querymany(ensembl_ids,
                   scopes='ensembl.gene',
                   fields='symbol',
                   species='human')

id2symbol = {
    r['query']: r.get('symbol')
    for r in res
    if not r.get('notfound', False)
}

df['Gene_symbol'] = df['Gene'].map(id2symbol)
df = df.loc[:, ~df.columns.str.contains('229E')]
```

```python
df = df.dropna(subset=['Gene_symbol'])
df = df.drop(columns=['Gene'])

cols = ['Gene_symbol'] + [col for col in df.columns if col != 'Gene_symbol']
df = df[cols]

# Using OC43 data only
df = df.loc[:, ~df.columns.str.contains('229E')]

print(df.head())
```

```python
df.to_csv("./CSV/Bulk-seq/GSE238079_converted.csv", index=False)
```
